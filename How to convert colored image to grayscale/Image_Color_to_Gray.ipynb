**AIM**: To create a program to display grayscale image.

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow
```

```python
# Function to show image and its size, shape, etc.
def img_color_to_grey(image,mode):
  cv2_imshow(image)
  print(f"Image Mode:{mode}")
  print(f"Image Type:{type(image)}")
  print(f"Image Size:{image.size}")
  print(f"Image Shape:{image.shape}")
```

```python
img_path ='/content/images.jpeg'
# print(img_path)
```

```python
colored_img = cv2.imread(img_path,cv2.IMREAD_COLOR)
# Reading the image as colored
```

```python
print(colored_img)
# image as numpy array
```

```python
if colored_img is None:
  print("Unable to read the input image")
else:
  img_color_to_grey(colored_img,"Colored Image")
  # Read image as grayscale
  gray_img = cv2.imread(img_path,cv2.IMREAD_GRAYSCALE)
  img_color_to_grey(gray_img,"GrayScale Image")
```
