**AIM:** To create a vision program to find histogram value and display histograph of a grayscale and color image.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
path = "/content/images.jpeg"
```

**For Gray Image:**

```python
gray_img = cv2.imread(path,cv2.IMREAD_GRAYSCALE)
```

```python
cv2_imshow(gray_img)
```

*Using Matplotlib:*

```python
plt.subplot(221),plt.imshow(gray_img,'gray')
plt.subplot(222),plt.hist(gray_img.flatten(),256,[0,256])
plt.show()
```

*Using OpenCv:*

```python
hist = cv2.calcHist([gray_img],[0],None,[256],[0,256])
plt.subplot(223),plt.imshow(gray_img,'gray')
plt.subplot(224),plt.plot(hist)
plt.xlim([0,256])
plt.show()
```

**For Colored Image:**

```python
colored_img = cv2.imread(path,cv2.IMREAD_COLOR)
```

```python
cv2_imshow(colored_img)
```

```python
color = ['r','g','b']
plt.subplot(223),plt.imshow(colored_img)
for i,col in  enumerate(color):
  hist2 = cv2.calcHist([colored_img],[i],None,[256],[0,256])
  plt.subplot(224),plt.plot(hist2,color = col)
plt.show()
```
